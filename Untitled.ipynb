```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt, math, pathlib

DATA_DIR = pathlib.Path(r"C:\Users\zmetc\OneDrive\Desktop\DSSR HIVTAR Analysis Project\DSSR output\torsions")
WINDOW = (19, 43)
ANGLES = ["alpha","beta","gamma","delta","epsilon","zeta","chi","phase-angle"]

def circ_mean_deg(vals):
    vals = pd.Series(vals).dropna().astype(float) % 360
    if len(vals)==0: return np.nan
    C,S = np.cos(np.deg2rad(vals)).mean(), np.sin(np.deg2rad(vals)).mean()
    return (np.degrees(np.arctan2(S,C)) + 360) % 360

def circ_sd_deg(vals):
    vals = pd.Series(vals).dropna().astype(float) % 360
    if len(vals)==0: return np.nan
    C,S = np.cos(np.deg2rad(vals)).mean(), np.sin(np.deg2rad(vals)).mean()
    R = np.hypot(C,S)
    return np.degrees(np.sqrt(-2*np.log(max(R,1e-12)))) if R>0 else 180.0
```

```python
csvs = list(DATA_DIR.glob("*-torsions_clean.csv"))
df = pd.concat([pd.read_csv(p) for p in csvs], ignore_index=True)
df.columns = [c.strip().lower() for c in df.columns]

df = df[(df["resi"]>=WINDOW[0]) & (df["resi"]<=WINDOW[1])]
df.head()
```

look at which residues have big SDs  those are flexible

```python
def agg_stats(group):
    out = {}
    for a in ANGLES:
        out[f"{a}_mean"] = circ_mean_deg(group[a])
        out[f"{a}_sd"] = circ_sd_deg(group[a])
    return pd.Series(out)

pooled = df.groupby(["resi"]).apply(agg_stats).reset_index()
pooled.to_csv(DATA_DIR/"pooled_circ_stats.csv", index=False)
pooled.head()
```

```python
mat = pooled.set_index("resi")[[f"{a}_sd" for a in ANGLES]]
plt.figure(figsize=(8,5))
plt.imshow(mat.T, aspect="auto", cmap="viridis")
plt.yticks(range(len(ANGLES)), ANGLES)
plt.xticks(range(len(mat.index)), mat.index)
plt.colorbar(label="circular SD (deg)")
plt.title("Torsional Flexibility Heatmap (residues 1943)")
plt.tight_layout()
plt.savefig(DATA_DIR / "flexibility_heatmap.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
sd_cols = [c for c in pooled.columns if c.endswith('_sd')]

pooled["max_sd"] = pooled[sd_cols].max(axis=1)
pooled["top_angle"] = pooled[sd_cols].idxmax(axis=1).str.replace('_sd','')

single_model_summary = pooled[["resi","max_sd","top_angle"]].sort_values("resi")
single_model_summary.to_csv(DATA_DIR / "single_model_summary.csv", index=False)

single_model_summary.head(10)
```

```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt, math

# Inputs
summary = pd.read_csv(DATA_DIR / "single_model_summary.csv")   # has resi, max_sd, top_angle
df_full = df.copy()                                            # from earlier (raw combined torsions)
THRESH = 35            # only show residues with max_sd >= THRESH (set to 0 for all)
TOP_N = None           # or set to an integer (e.g., 16) to take top-N flexible residues

# Select which residues to plot
subset = summary.query("max_sd >= @THRESH").sort_values("max_sd", ascending=False)
if TOP_N is not None:
    subset = subset.head(TOP_N)

# Choose grid size
n = len(subset)
if n == 0:
    raise ValueError("No residues match the selection. Lower THRESH or set TOP_N.")

COLS = 4
ROWS = math.ceil(n / COLS)

# Helper: draw one rose histogram on an axis
def rose_on_axis(ax, angles_deg, bins=24):
    if len(angles_deg) == 0:
        ax.text(0.5, 0.5, "No data", ha="center", va="center", transform=ax.transAxes)
        return
    theta = np.deg2rad(np.array(angles_deg) % 360.0)
    edges = np.linspace(0, 2*np.pi, bins + 1)
    counts, _ = np.histogram(theta, bins=edges)
    centers = (edges[:-1] + edges[1:]) / 2
    widths  = np.diff(edges)
    ax.bar(centers, counts, width=widths, align="center")
    ax.set_yticklabels([])   # cleaner look
    ax.set_xticklabels([])

# Build the figure
fig = plt.figure(figsize=(COLS*3, ROWS*3))
for i, (resi, max_sd, angle) in enumerate(zip(subset["resi"], subset["max_sd"], subset["top_angle"])):
    vals = df_full.query("resi == @resi")[angle].dropna().astype(float).values
    ax = plt.subplot(ROWS, COLS, i+1, projection="polar")
    rose_on_axis(ax, vals, bins=24)
    ax.set_title(f"Res {resi}  {angle}  (SD{max_sd:.1f})", fontsize=9, pad=10)

# Fill any unused panels with blanks
for j in range(i+1, ROWS*COLS):
    ax = plt.subplot(ROWS, COLS, j+1, projection="polar")
    ax.axis("off")

plt.suptitle("Rose plots  most flexible torsion per residue", y=0.98, fontsize=12)
plt.tight_layout(rect=[0,0,1,0.97])
plt.savefig(DATA_DIR / "rose_contact_sheet_top_angle.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
ANGLE = "chi"            # change to "epsilon", "gamma", etc.
TOP_N = 16               # show top-N residues ranked by SD for this angle

# Rank residues by SD of this fixed angle
sd_col = f"{ANGLE}_sd"
ranked = (pooled[["resi", sd_col]]
          .dropna()
          .sort_values(sd_col, ascending=False)
          .head(TOP_N))

n = len(ranked)
COLS = 4
ROWS = math.ceil(n / COLS)

fig = plt.figure(figsize=(COLS*3, ROWS*3))
for i, (resi, sd) in enumerate(zip(ranked["resi"], ranked[sd_col])):
    vals = df_full.query("resi == @resi")[ANGLE].dropna().astype(float).values
    ax = plt.subplot(ROWS, COLS, i+1, projection="polar")
    rose_on_axis(ax, vals, bins=24)
    ax.set_title(f"Res {resi}  {ANGLE}  (SD{sd:.1f})", fontsize=9, pad=10)

for j in range(i+1, ROWS*COLS):
    ax = plt.subplot(ROWS, COLS, j+1, projection="polar")
    ax.axis("off")

plt.suptitle(f"Rose plots  {ANGLE} across residues", y=0.98, fontsize=12)
plt.tight_layout(rect=[0,0,1,0.97])
plt.savefig(DATA_DIR / f"rose_contact_sheet_{ANGLE}.png", dpi=300, bbox_inches="tight")
plt.show()
```
